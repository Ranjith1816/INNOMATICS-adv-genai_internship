 Problem 3: Student Result Processing System

```python
student_marks = [90, 75, 50, 60]
def evaluation(marks):
  total=0
  for mark in marks:
    total += mark
  average =total/len(marks)
  if average>=50:
      return "Pass"
  else:
      return "Fail"
avg=sum(student_marks) /len(student_marks)
result=evaluation(student_marks)
print("Average:",avg)
print("Result:",result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12

    # Calculate normal fare
    fare = base_fare + (distance * per_km_rate)

    # Add peak hour charge (25%)
    if is_peak:
        fare += fare * 0.25

    return fare


# Retry system using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        if peak_input not in ["yes", "no"]:
            print("Please enter 'yes' or 'no'.\n")
            continue

        is_peak = True if peak_input == "yes" else False

        total_fare = calculate_fare(distance, is_peak)

        print(f"Total Fare: ₹{total_fare:.2f}")
        break

    except ValueError:
        print("Invalid input. Please enter numeric distance.\n")
```

Problem 5: Employee Attendance Eligibility Checker

```python
attendance_list = ["P", "A", "P", "A", "A", "P", "P", "A"]
def eligibility(attendance):
  totaldays=len(attendance)
  presentdays=0
  for status in attendance:
    if status == "P":
      presentdays +=1
  percentage = (presentdays / totaldays) * 100
  if percentage >=75:
    return percentage , "ELIGIBLE"
  else:
    return percentage , "NOT ELIGIBLE"
result = eligibility(attendance_list)

print("Attendance Result:", result)
```

Problem 6: Password Strength Checker

```python
def pwd_strength(password):
    has_digit = False
    has_special = False
    if len(password) < 8:
        return "password must be atleast eight characters"
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
user_password = input("Enter password: ")
result = pwd_strength(user_password)
print(result)
```
