Problem Statement 1:Smart Parking Lot Management System


```python
def smart_parking_system(capacity, vehicle_logs):
    current_count = 0
    peak_count = 0

    for log in vehicle_logs:
        if log == "IN":
            current_count += 1
        else:
            current_count -= 1

        # Track peak usage
        if current_count > peak_count:
            peak_count = current_count

    print("Currently Parked Vehicles:", current_count)


    if current_count > capacity:
        print("ALERT: Parking capacity exceeded!")
    elif current_count == capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)

```

Problem Statement 2:Online Food Delivery Time Estimator


```python

def estimate_delivery_time(distance, traffic_level, weather):

    # Base speed assumption (km/h)
    average_speed = 30

    # Base time in minutes
    base_time = (distance / average_speed) * 60

    # Traffic multipliers
    traffic_multipliers = {
        "Low": 1.0,
        "Medium": 1.2,
        "High": 1.4
    }

    # Weather multipliers
    weather_multipliers = {
        "Clear": 1.0,
        "Rainy": 1.25,
        "Storm": 1.5
    }

    # Apply conditions
    adjusted_time = base_time
    adjusted_time *= traffic_multipliers.get(traffic_level, 1.0)
    adjusted_time *= weather_multipliers.get(weather, 1.0)
    print("Estimated Delivery Time:", round(adjusted_time), "minutes")
estimate_delivery_time(8, "High", "Rainy")

```

Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def analyze_seat_booking(total_seats, booked_seats):

    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional show")
    else:
        print("Show Status: Seats Available")
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

analyze_seat_booking(total_seats, booked_seats)
```

Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):

    # Calculate total using loop
    total = 0
    for value in cpu_readings:
        total += value

    # Calculate average
    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def track_resource_usage(resource_usage):

    threshold = 8   # hours limit
    overused_resources = []

    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused_resources.append(resource)

    # Output results
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

track_resource_usage(usage_data)
```

Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_event_registrations(capacity, registrations):

    confirmed = 0
    waitlisted = 0

    # Loop through registrations
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")
manage_event_registrations(100, 105)
```
