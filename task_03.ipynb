Problem Statement 1: Employee Performance Bonus Eligibility


```python


employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

#Find highest performance score
highest_score = max(employees.values())

#Find all employees with highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)


print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {highest_score})")
```

Problem Statement 2: Search Query Keyword Analysis

```python
# Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
import string
for char in string.punctuation:
    query = query.replace(char, "")

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Step 5: Display only words searched more than once
repeated_words = {}

for word, count in word_count.items():
    if count > 1:
        repeated_words[word] = count

print(repeated_words)
```

Problem Statement 3: Sensor Data Validation

```python
# Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Loop through readings with index
for hour, reading in enumerate(sensor_readings):

    # Check if reading is even (valid)
    if reading % 2 == 0:
        valid_readings.append((hour, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis

```python
# Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Step 1: Extract and count domains
for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 2: Calculate percentage usage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

Problem Statement 5: Sales Spike Detection

```python
# Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)
# Step 3: Detect spikes
for day, value in enumerate(sales, start=1):
    if value >average_sales :
        print(f"Day {day}: {value}")
```

Problem Statement 6: Duplicate User ID Detection


```python
# Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Step 1: Count occurrences
for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

# Step 2: Display duplicates
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
